```python
# Importing necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# importing sklearn libraries

from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
```

We are going to import our dataset and seeing if there is any missing information as well as the type of data and type of the features

```python
mark_data = pd.read_csv('bank.csv', sep=';')

# Checking the first 5 rows of our dataset
print(mark_data.head())

# A more in-depth look at our data
print(mark_data.describe())
```

Let's now star our EDA by checking any correlation between numeric columns (pair plot)

```python
# Select only numerical columns
numerical_cols = mark_data.select_dtypes(include=['int64', 'float64']).columns

# Create the pair plot for a 500 sample
sns.pairplot(mark_data[numerical_cols].sample(500, random_state=42))

# Show the plot
plt.show()
plt.clf()
```

As we could notice, it seems that our data samples have some outliers, so we are going to check these, by ploting box plots for these numeric columns

```python
# Setting the plot size
plt.figure(figsize=(12, 8))

# Creating boxplots for each numerical column
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 3, i)  # Arranging plots in a grid
    sns.boxplot(y=mark_data[col])
    plt.title(col)

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
plt.clf()
```

Great! Now that we see the huge difference in results, we need to start standardizing our columns as well as defining our categorical columns so we can start creating our pipeline appropriately.

```python
# Choosing our X and y values
X = mark_data.drop(columns=['y'])
y = mark_data['y']

# Selecting only the categorical features (columns)
categorical_cols = X.select_dtypes(include=['object']).columns
# print(categorical_cols) checking our columns

# Splitting data for our ML algorithm
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size = .2)
```

Now, let's start Pipelining!

```python
# First let's define our numerical columns for our training data
x_train_num = x_train[numerical_cols]

# Define our scaler (ChatGPT's recommendation for columns with too many outliers was RobustScaler instead of StandardScaler)
num_transformer = Pipeline([('scale', RobustScaler())])

# Defining transformer for categorical columns
cat_transformer = Pipeline([('ohe', OneHotEncoder(sparse_output=False, handle_unknown='ignore'))])

# Create column transformer that will preprocess the numerical and categorical features separately
preprocess = ColumnTransformer( transformers=[("cat_process", cat_transformer, categorical_cols), ("num_process", num_transformer, numerical_cols) ] )
```

Let's now combine one step of our EDA (PCA) with our last pipeline, using our column transformer

```python
# Creating a pipeline with preprocess, PCA, and a logistic regresssion model
pipeline = Pipeline([('preprocess', preprocess), ('pca', PCA(n_components=25)), ('clf', LogisticRegression(solver='saga', max_iter=10000))])

# Fit the pipeline on the training data
pipeline.fit(x_train, y_train)

#Predict the pipeline on the test data
y_pred = pipeline.score(x_test, y_test)
print('Test accuracy: ')
print(y_pred)

# Define search space of hyperparameters
search_space = [{'clf':[LogisticRegression(solver='saga', max_iter=10000)],
                     'clf__C': np.logspace(-3, 1, 5),
                'pca__n_components':np.linspace(20,25,3).astype(int)},
                   ]

# Search over hyperparameters above to optimize pipeline and fit, using GridSearchCV()

gs = GridSearchCV(pipeline, search_space, cv=5)
gs.fit(x_train, y_train)
```

Great! Now it looks like our model is performing really well, even before testing our hyperparameter tuning.
Let's try our GridSearchCV() best estimator and compare results!

```python
# Save the best estimator from the gridsearch and print attributes and final accuracy on test set
best_model = gs.best_estimator_

# Print attributes of best_model
print('The best classification model is:')
print(best_model.named_steps['clf'])
print('The hyperparameters of the best classification model are:')
print(best_model.named_steps['clf'].get_params())
print('The number of components selected in the PCA step are:')
print(best_model.named_steps['pca'].n_components)

# Print final accuracy score 
print(best_model.score(x_test, y_test))
```
